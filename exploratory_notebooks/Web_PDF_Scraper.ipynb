```python
import requests
import fitz
```

```python
def download_pdf(url, save_path):


    try:
        # Pass the headers with the GET request
        response = requests.get(url, stream=True, headers=headers)

        # Raise an exception if the download fails
        response.raise_for_status()

        # Open the local file in write-binary mode and save the content
        with open(save_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f"Successfully downloaded PDF to '{save_path}'")
        return True
    except requests.exceptions.RequestException as e:
        print(f"Error downloading PDF: {e}")
        return False
```

```python
# URL
pdf_url = "https://www.rand.org/content/dam/rand/pubs/perspectives/PE200/PE237/RAND_PE237z1.arabic.pdf"

# path
local_pdf_path = "downloaded_document.pdf"

# call the function to download the file
download_pdf(pdf_url, local_pdf_path)
```

```python
pdf_path = "downloaded_document.pdf"
```

```python
print("--- Extracting text with PyMuPDF ---")
try:
    doc = fitz.open(pdf_path)  # Open the PDF document
    full_text_pymupdf = ""     # Initialize string for extracted text

    for page in doc:           # Iterate through each page
        full_text_pymupdf += page.get_text() # Extract text and append

    doc.close()                # Close the PDF document

    print("Extraction successful. First 500 characters:")
    print(full_text_pymupdf[:5000]) # Print first 5000 characters of extracted text

except Exception as e:
    print(f"An error occurred with PyMuPDF: {e}") # Handle any errors

print("\n" + "="*50 + "\n")
```
